```python
import matplotlib.pyplot as plt
import numpy as np
from mathutils import *



# 2. Define the quadratic equation (y = x^2)
def quadratic(coeff=[]) :
    assert(len(coeff)==3)
    a = coeff[0]
    b = coeff[1]
    c = coeff[2]
    print(f"a={a}, b={b}, c={c}")
    return lambda x: a*x**2 + b*x + c

# Enter coefficients for the quadratic ax^2 + bx + c
a = -2
b = 20
c = -48

coefficients = [a,b,c]

quadratic_function = quadratic(coefficients) # create the function
title2=generate_poly_title(coefficients)
print(title2)


h =int( -(b / (2 * a)) )
k = int(quadratic_function(h))

hk = f"h,k = ({h}, {k})"

print(hk)

# 1. Generate x-values (witin a range 10, creating 400 points in between)
x = np.linspace(h-5, h+5, 400)
y = quadratic_function(x) ## get the Y-values

for xv in range(int(h-5),int(h+5)):
    yv=quadratic_function(xv)
    rec = f"| {xv} | {yv} |"
    if xv == h and yv == k:
        rec += "-----"
    print(rec)
        
    



# 3. Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=title2, color="blue", linewidth=2)

# 4. Customize the graph to make it readable
plt.title(title2, fontsize=14)
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # X-axis line
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Y-axis line
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend()

# 5. Display the graph
plt.show()
```
