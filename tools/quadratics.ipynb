```python
import matplotlib.pyplot as plt
import numpy as np
from mathutils import *
import csv

# 2. Define the quadratic equation (y = x^2)
def quadratic(coeff=[]) :
    assert(len(coeff)==3)
    a = coeff[0]
    b = coeff[1]
    c = coeff[2]
    print(f"a={a}, b={b}, c={c}")
    return lambda x: a*x**2 + b*x + c

# Enter coefficients for the quadratic ax^2 + bx + c
a = -5
b = 2
c = -3

# create the function
coefficients = [a,b,c]
quadratic_function = quadratic(coefficients) # create the function
title2=generate_poly_title(coefficients)

# find the vertex P(x) = (x - h)^2 + k
h =int( -(b / (2 * a)) )
k = int(quadratic_function(h))

hk = f"h,k = ({h}, {k})"
print(hk+"/n/n")



## verify points around the vertex in view_values.csv
points_verify=[[h,k,"***"]]
L = []
H = []
range_l=0
range_h=0
for i in range(1,5):
    L.append([h-i,int(quadratic_function(h-i)),None])
    H.append([h+i,int(quadratic_function(h+i)),None])
writer = csv.writer(open("view_values.csv",'w'))
writer.writerow(["X","Y","vertex"])
points_verify  = list(reversed(L)) + points_verify + H
range_l = points_verify[0][0]
range_h = points_verify[-1][0]
for p in points_verify: writer.writerow(p)
writer = None

# Generate the graph data

x = np.linspace(range_l, range_h, 400)
y = quadratic_function(x) ## get the Y-values

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=title2, color="blue", linewidth=2)

# 4. Customize the graph to make it readable
plt.title(title2, fontsize=14)
plt.xlabel(f"X Axis     {hk} ", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # X-axis line
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Y-axis line
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend()

# 5. Display the graph
plt.show()
```
